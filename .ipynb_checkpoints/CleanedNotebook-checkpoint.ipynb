```python
#Importing tools
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from scipy.optimize import minimize
```

```python
#reading the data
df = pd.read_csv('hlx.csv')
df
```

```python
#function to  return a date range that is requested
import datetime
def year_range(df, start_date, end_date):
    y_range = (df['Date'] > start_date) & (df['Date'] <= end_date)
    return y_range 
```

```python
#seperating the data for the year range wanted
df['Date'] = pd.to_datetime(df['Date']).dt.date
y_range = year_range(df, datetime.date(2015,1,1), datetime.date(2018,1,1))

c_df = df.loc[y_range]
c_df
```

```python
#further cleaning the data
c_df.index = pd.to_datetime(c_df['Date'])
c_df

c_df = c_df.drop(['Date'], axis=1)
c_df
```

```python
#unncessary just so it could be easier to read
X = c_df

#creating a new column that would be used for identifying the target labels.
X['open-close'] = X['Open'] - X['Close']
X['high-low'] = X['High'] - X['Low']
X
```

```python
#creating the target labels
T = np.where(X['open-close'] > 0, 1, -1)
T
```

```python
#function to split the data (from labs)
from sklearn.model_selection import train_test_split

def data_splitting(data, target):
    
    X_train, X_test, t_train, t_Test = train_test_split(data, target, random_state=0, test_size=0.20)
    return X_train, X_test, t_train, t_Test
```

```python
#function for kernel found from kaggle
def kernel(x,y):
    sigma = 1.0
    return np.exp(-np.linalg.norm(x-y)**2/(2*sigma**2))
```

```python
#custom svm fit function found from kaggle
def fit_svm(X, y, C):
    N = len(y)
    bounds = [(0, C) for i in range(N)]
    alpha0 = np.zeros(N)
    res = minimize(objective_function, alpha0, args=(y, X, C), bounds=bounds)
    alpha = res.x
    w = np.sum(alpha*y[:,np.newaxis]*X, axis=0)
    b = np.mean(y - X.dot(w))
    return w, b
```

```python
#custom objective function found from kaggle
def objective_function(alpha, y, X, c):
    N = len(y)
    K = np.zeros((N, N))
    for i in range(N):
        for j in range(N):
            K[i,j] = y[i] * y[j] * kernel(X[i], X[j])
    s = 0
    for i in range(N):
        if 0 < alpha[i] < C:
            s += alpha[i]*y[i]
    return -np.sum(alpha) + 0.5*np.sum(alpha*K.dot(alpha)) - s
```

```python
#checking to see if split was done correctly (from labs)
X_train, X_test, t_train, t_test = data_splitting(X, T)

print("Train data shape: {}".format(X_train.shape))
print("Train target shape: {}".format(t_train.shape))
print("Test data shape: {}".format(X_test.shape))
print("Test target shape: {}".format(t_test.shape))
```

```python
#training the model and evaluating the potential accuracy
svm = SVC(kernel='rbf', C=1, gamma='scale')
svm.fit(X_train, t_train)

train_score = svm.score(X_train, t_train)
test_score = svm.score(X_test, t_test)

svm_train = svm.predict(X_train)
svm_test = svm.predict(X_test)

print(f"Train Accuracy: {train_score}\nTest Accuracy: {test_score}")
```

```python
#obtaining the prediction values from the close feature in data set 
stock_predict = svm.predict(X)
```

```python
#calculating the predicted daily change
cal_daily_close = X['Close'].pct_change() * stock_predict
```
